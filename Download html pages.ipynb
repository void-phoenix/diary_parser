```python
import os
from time import sleep
from bs4 import BeautifulSoup
import requests
import numpy as np
import shutil
from pathlib import Path
```

```python
base_path = 'C:\\Users\\void\\conda projects\\diary_parser\\diaries'
user_folders = [f for f in os.listdir(base_path) if not f.startswith('.')]
```

```python
def get_user_links(user_folder):
    links = []
    with open(base_path + "\\" + user_folder + "\\links") as f:
        for link in f.readlines():
            links.append(link)
        
    links = [l[:-1] for l in links]
    return links
```

```python
for folder in user_folders:
    links = get_user_links(folder)
    pages_folder = base_path + "\\" + folder + "\\pages"
    if not (Path(pages_folder).exists()):
        os.makedirs(pages_folder)
    was_processed = False
    for link in links:
        page_id = link.split('/')[-1]
        file_name = (pages_folder + "\\" + page_id)
        if (Path(file_name).exists()):
            continue
        page = requests.get(link)
        html = page.text
        try:
            with open(file_name, 'w') as f:
                f.write(str(html.encode("utf8")))
            was_processed = True
        except Exception :
            print(file_name)
    if was_processed:
        print(folder)
```
